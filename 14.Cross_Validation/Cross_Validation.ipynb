**Cross Validation**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold,cross_val_score # K-fold cross validation
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, SCORERS
```

```python
data=pd.read_csv("E:\\AI & Machine Learning\\03.Machine Learning\\Practice\\Datasets-CSV\\Boston.csv")
```

```python
data.head(10)
```

```python
x=data.iloc[:,:12].values
y=data.iloc[:,12].values
```

```python
kf=KFold(n_splits=5) # 5-fold cross validation
```

```python
model=LinearRegression() # Linear Regression model,
```

```python
list(kf.split(data)) #output = [(train_index, test_index), (train_index, test_index), (train_index, test_index), (train_index, test_index), (train_index, test_index)]
```

```python
RMSE=[]

for i in list(kf.split(data)): # i = (train_index, test_index)
    model.fit(x[i[0]],y[i[0]]) # train model
    y_pred=model.predict(x[i[1],:]) # test model
    RMSE.append(np.sqrt(mean_squared_error(y_pred,y[i[1]]))) # calculate RMSE
    
```

```python
RMSE # RMSE for each fold
```

```python
np.mean(RMSE) # average RMSE
```

onother methord

```python
cross_val_score(model,x,y,cv=10,scoring="neg_mean_squared_error") # 10-fold cross validation
```

```python
cvals=-cross_val_score(model,x,y,cv=10,scoring="neg_mean_squared_error") # RMSE for each fold
```
