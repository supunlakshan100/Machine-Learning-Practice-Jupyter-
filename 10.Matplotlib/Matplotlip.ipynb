```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
L1=[1,2,3,4,5]
L2=[10,25,32,58,65]
plt.plot(L1,L2)
```

```python
L1=[1,2,3,4,5]
L2=[10,25,32,58,65]
plt.plot(L1,L2)
plt.title("A simple line chart")
plt.xlabel("L1")
plt.ylabel("L2")
```

```python
L1=[1,2,3,4,5]
L2=[10,25,32,58,65]

plt.figure(figsize=(5,3))

plt.plot(L1,L2,"red")
plt.title("A simple line chart")
plt.xlabel("L1")
plt.ylabel("L2")
plt.show()
```

```python
L1=[1,2,3,4,5]
L2=[10,25,32,58,65]

plt.figure(figsize=(5,3))

plt.plot(L1,L2,"ro")
plt.title("A simple line chart")
plt.xlabel("L1")
plt.ylabel("L2")
plt.show()
```

```python
L1=[1,2,3,4,5]
L2=[10,25,32,58,65]

plt.figure(figsize=(5,3))

plt.plot(L1,L2,"b*")
plt.title("A simple line chart",color="red")
plt.xlabel("L1",color="red")
plt.ylabel("L2",color="red")
plt.show()
```

Pie chart

```python
Gender=["Male","Female"]
Number_of_Students=[45,75]

plt.pie(Number_of_Students,labels=Gender,autopct="%0.1f")
plt.legend()
plt.show()
```

```python
Gender=["Male","Female"]
Number_of_Students=[45,75]

plt.figure(figsize=(8,8))

plt.pie(Number_of_Students,labels=Gender,autopct="%0.1f%%")
plt.legend()
plt.show()
```

Scattter plots

```python
height=np.random.normal(160,5,100)
weight=np.random.normal(55,2,100)

plt.scatter(height,weight)
plt.title("Height vs Weight",color="red")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
plt.scatter(df["Avg. Area Number of Rooms"],df["Price"],color="b",marker="+")
plt.title("Avg. Area Number of Rooms vs Price",color="b",style="italic")
plt.xlabel("Avg. Area Number of Rooms")
plt.ylabel("Price")
plt.show()
```

```python
plt.scatter(df["Avg. Area Number of Rooms"],df["Price"],c="g",marker="+")
plt.title("Avg. Area Number of Rooms vs Price",c="g",style="italic")
plt.xlabel("Avg. Area Number of Rooms")
plt.ylabel("Price")
plt.show()
```
