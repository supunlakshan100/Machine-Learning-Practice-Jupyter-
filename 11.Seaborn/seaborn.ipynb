```python
Titanic=sns.load_dataset('titanic')
Titanic.head()
```

```python
sns.displot(Titanic['age'],kde=True)
plt.show()
```

```python
sns.boxplot(Titanic["age"])
plt.show()
```

```python
df=pd.DataFrame(Titanic.groupby('sex')["sex"].count())
df.columns=['Count']
df.reset_index(inplace=True)
```

```python
df
```

```python
sns.barplot(data=df,x='sex',y="Count")
plt.show()
```

```python
Titanic.head()
```

```python
sns.catplot(data=Titanic,x='sex',kind='count',hue='alive')
plt.figure(figsize=(10,10))
plt.show()
```

```python
sns.scatterplot(data=Titanic,x='pclass',y='age',hue='alive')
plt.show()
```

```python
df2=pd.DataFrame({"Loc1":[0,10,100],"Loc2":[10,0,100],"Loc3":[100,100,0]},index=['Loc1','Loc2','Loc3'])
```

```python
df2
```

```python
sns.heatmap(df2,annot=True)
plt.xlabel('Location')
plt.ylabel('Location')
plt.title('Heatmap',color='red',style='italic')
plt.show()
```
