```python
import numpy as np
import pandas as pd
```

```python
df1=pd.DataFrame(np.arange(75).reshape(15,5) ,columns=('col1','col2','col3','col4','col5')) # 15 rows and 5 columns
df1
```

```python
df1.apply(sum, axis=0) # axis=0 means column wise
```

```python
df1.apply(sum, axis=1) # axis=1 means row wise, most suitable for row wise operation
```

```python
fun=lambda x: x.max()-x.min()
df1.apply(fun, axis=0) # column wise max-min
```

```python
fun=lambda x: x.max()-x.min()
df1.apply(fun, axis=1) # row wise max-min
```

```python
df1['col1'].map(lambda arr: arr+1) # add 1 to each element of col1
```

```python
fun=lambda x: x+100
```

```python
df1.apply(fun)
```

```python
df1['col1'].map(fun) # add 1 to each element of col1
```

```python
df1['col2']>20 # boolean output
```

```python
df1[df1['col2']>20]  # filter rows where col2>20
```

```python
df1[df1['col2']<70]
```
